```python
import pandas as pd
import numpy as np
```

```python
#load excel into dataframe
df = pd.read_csv(r'C:\Users\jyama\project_stuff\weed_stuff\Medical_Marijuana_Brand_Registry.csv')
names = pd.DataFrame(df['BRAND-NAME'])
df = df.drop(['BRAND-NAME'], axis = 1)
column_names = list(df.columns)

df.head()
```

```python
# nan values, in this case, represent undetected amounts of a certain substance. therefore nan = o
df = df.fillna(0)
df.head()
```

```python
print(list(df.dtypes))
```

```python
df = df.apply(pd.to_numeric, errors='coerce', downcast='float')
```

```python
print(list(df.dtypes))
```

```python
df.shape
```

```python
df = df.dropna()
df.shape
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['TETRAHYDROCANNABINOL-ACID-THCA'],df['CANNABIDIOL-ACID-CBDA'])
plt.xlabel('THCa')
plt.ylabel('CBDa')
```

```python
#remove outliers from data
from scipy import stats
z = np.abs(stats.zscore(df))

# set threshold for z-scores
threshold = 5.5

# identify and remove rows that have z-scores greater than the threshold
df = df[(z < threshold).all(axis=1)]
df.shape
```

```python
#normalize data
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = MinMaxScaler()
scaler.fit(df)
df_mm = scaler.transform(df)
df_mm = pd.DataFrame(df_mm)

df_mm.rename(columns=dict(zip(df_mm.columns, column_names)), inplace=True)
df_mm.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df_mm['TETRAHYDROCANNABINOL-ACID-THCA'],df_mm['CANNABIDIOL-ACID-CBDA'])
```

```python
from sklearn.cluster import KMeans
k = range(1,5)
sse = []
for i in k:
    km = KMeans(n_clusters =i)
    km.fit(df)
    sse.append(km.inertia_)
plt.plot(k,sse)
```

```python
#indica vs sativa 
# 2 clusters
from sklearn.model_selection import KFold
from sklearn.metrics import silhouette_score

n_folds = 5
kf = KFold(n_splits = n_folds)

scores = []

k = 2

for train_index, test_index in kf.split(df_mm):
    # Get the training and testing data for this fold
    X_train, X_test = df_mm.iloc[train_index], df_mm.iloc[test_index]

    # Fit a k-means model on the training data
    kmeans = KMeans(n_clusters=k).fit(X_train)

    # Predict the cluster assignments on the test data
    y_pred = kmeans.predict(X_test)

    # Compute the silhouette score on the test data
    score = silhouette_score(X_test, y_pred)

    # Append the score to the list of scores
    scores.append(score)

# Compute the mean and standard deviation of the silhouette scores
mean_score = np.mean(scores)
std_score = np.std(scores)
print(mean_score,std_score)
```

```python
#perform pca to reduce dimensions
from sklearn.decomposition import PCA
n_components = range(1,15)
pca = PCA()
pca_scores = [pca.fit(df_mm).explained_variance_ratio_.cumsum()[i-1] for i in n_components]

# Plot the explained variance ratio for each value of n_components
plt.plot(n_components, pca_scores)
plt.xlabel('Number of components')
plt.ylabel('Explained variance ratio')
plt.show()
```

```python
pca = PCA(n_components=5)

# Fit the PCA model to the data
pca.fit(df_mm)

# Transform the data using the PCA model
df_pca = pca.transform(df_mm)
df_pca = pd.DataFrame(df_pca)
df_pca.head()
```

```python
n_folds = 5
kf = KFold(n_splits = n_folds)

scores = []

k = 2
for train_index, test_index in kf.split(df_pca):
    # Get the training and testing data for this fold
    X_train, X_test = df_pca.iloc[train_index], df_pca.iloc[test_index]

    # Fit a k-means model on the training data
    kmeans = KMeans(n_clusters=k).fit(X_train)

    # Predict the cluster assignments on the test data
    y_pred = kmeans.predict(X_test)

    # Compute the silhouette score on the test data
    score = silhouette_score(X_test, y_pred)

    # Append the score to the list of scores
    scores.append(score)

# Compute the mean and standard deviation of the silhouette scores
mean_score = np.mean(scores)
std_score = np.std(scores)
print(mean_score,std_score)  
```
